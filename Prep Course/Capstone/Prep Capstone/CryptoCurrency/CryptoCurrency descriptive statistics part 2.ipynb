```python
# Crypto Currency descriptive statistics
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import datetime
%matplotlib inline
```

```python
crp = pd.read_csv('crypto-markets.csv')
```

```python
crp['date'] = pd.to_datetime(crp['date'], format = '%Y-%m-%d', utc=True)
```

```python
# Which crypto Currencies are valued the highest?

crp_low = crp[(['ranknow', 'name', 'low'])]
crp_low = crp_low.groupby('name').mean()
crp_low = crp_low.sort_values('ranknow')
crp_low = crp_low['low']

crp_high = crp[(['ranknow', 'name', 'high'])]
crp_high = crp_high.groupby('name').mean()
crp_high = crp_high.sort_values('ranknow')
crp_high = crp_high['high']

ax = crp_low.head(40).plot(kind='bar', title='Crypto Low Values', figsize=(15,5))
ax.set_xlabel("Rank")
ax.set_ylabel("Mean Dollar Value")
plt.show()

ax = crp_high.head(40).plot(kind='bar', title='Crypto High Values', figsize=(15,5))
ax.set_xlabel("Rank")
ax.set_ylabel("Mean Dollar Value")
plt.show()

# Which crypto currencies have the largest spread?

crp_spread = crp[(['ranknow', 'name', 'spread'])]
crp_spread = crp_spread.groupby('name').mean()
crp_spread = crp_spread.sort_values('ranknow')
crp_spread = crp_spread['spread']

ax = crp_spread.head(40).plot(kind='bar', title='Crypto Spread', figsize=(15,5))
ax.set_xlabel("Rank")
ax.set_ylabel("Mean Spread")
plt.show()
```

```python
# Which crypto currencies have had the highest growth based on market cap?

crpBTC = crp[crp['ranknow']==1]
crpBTC = crpBTC[(['date', 'market'])]
crpBTC = crpBTC[crpBTC['date']>='2018-01-01']
crpBTC = crpBTC.set_index('date')
bx = crpBTC.plot(figsize=(15,4), title='Bitcoin 2018 market Value Trend')
bx.set_xlabel('2018')
bx.set_ylabel('market Value')
plt.show()

crpETH = crp[crp['ranknow']==2]
crpETH = crpETH[(['date', 'market'])]
crpETH = crpETH[crpETH['date']>='2018-01-01']
crpETH = crpETH.set_index('date')
ax = crpETH.plot(figsize=(15,4), title='Ethereum 2018 market Value Trend')
ax.set_xlabel('2018')
ax.set_ylabel('market Value')
plt.show()

crpXRP = crp[crp['ranknow']==3]
crpXRP = crpXRP[(['date', 'market'])]
crpXRP = crpXRP[crpXRP['date']>='2018-01-01']
crpXRP = crpXRP.set_index('date')
ax = crpXRP.plot(figsize=(15,4), title='Ripple 2018 market Value Trend')
ax.set_xlabel('2018')
ax.set_ylabel('market Value')
plt.show()

crpBCH = crp[crp['ranknow']==4]
crpBCH = crpBCH[(['date', 'market'])]
crpBCH = crpBCH[crpBCH['date']>='2018-01-01']
crpBCH = crpBCH.set_index('date')
ax = crpBCH.plot(figsize=(15,4), title='Bitcoin Cash 2018 market Value Trend')
ax.set_xlabel('2018')
ax.set_ylabel('market Value')
plt.show()

crpEOS = crp[crp['ranknow']==5]
crpEOS = crpEOS[(['date', 'market'])]
crpEOS = crpEOS[crpEOS['date']>='2018-01-01']
crpEOS = crpEOS.set_index('date')
ax = crpEOS.plot(figsize=(15,4), title='EOS 2018 market Value Trend')
ax.set_xlabel('2018')
ax.set_ylabel('market Value')
plt.show()

# Add months to this data
```

```python
# Top 40 coin market caps

crp_market = crp[(['ranknow', 'name', 'market'])]
crp_market = crp_market.groupby('name').mean()
crp_market = crp_market.sort_values('ranknow')
crp_market = crp_market['market']

ax = crp_market.head(40).plot(kind='bar', title='Crypto Market Cap Values', figsize=(15,5))
ax.set_xlabel("Name")
plt.xticks(rotation='vertical')
ax.set_ylabel("Mean Market Cap")
plt.show()

# Some anomolies in market cap values i.e. why is "3" less than "4" or why is "8" so far back
# Is there a away to write the x labels horizontally?
```

```python
# Explore the outliers, Bit20, NANOX.

# Crypto Low Values for outliers
crp_low = crp[(['ranknow','name', 'low'])]
crp_low = crp_low.groupby('name').mean()
crp_low = crp_low.sort_values('ranknow')
crp_low = crp_low[['low']]
crp_MVP_low = crp_low[crp_low['low']>=105]

ax = crp_MVP_low.head(40).plot(kind='bar', title='Crypto MVP Low Values', figsize=(15,5))
ax.set_xlabel("Name")
ax.set_ylabel("Mean Dollar Value")
plt.show()

# Bit20 trend
crpBit20 = crp[crp['ranknow']==957]
crpBit20 = crpBit20[(['date', 'market'])]
crpBit20 = crpBit20[crpBit20['date']>='2018-01-01']
crpBit20 = crpBit20.set_index('date')
ax = crpBit20.plot(figsize=(15,4), title='Bit20 2018 Market Cap Trend')
ax.set_xlabel('2018')
ax.set_ylabel('Market Cap')
plt.show()

# NANOX trend
crpNANOX = crp[crp['ranknow']==1021]
crpNANOX = crpNANOX[(['date', 'market'])]
crpNANOX = crpNANOX[crpNANOX['date']>='2018-01-01']
crpNANOX = crpNANOX.set_index('date')
ax = crpNANOX.plot(figsize=(15,4), title='NANOX 2018 Market Cap Trend')
ax.set_xlabel('2018')
ax.set_ylabel('Market Cap')
plt.show()
# NANOX seems to be an odd coin. Can't find resources telling me exactly what it is.
```

```python
# Create some scatter plots comparing spread with market cap and close ratio. And two coins such as 
# Ethereum and Bitcoin.

crpBTC = crp[crp['ranknow']==1]
crpBTC = crpBTC[(['date', 'market', 'spread'])]
crpBTC = crpBTC[crpBTC['date']>='2018-01-01']
plt.scatter(x=crpBTC['spread'], y=crpBTC['market'])
plt.title('Spread vs. Market Cap')
plt.xlabel('Spread')
plt.ylabel('Market Cap')
            
crp2 = crp[crp['ranknow']<=5]
crp2 = crp2[(['date', 'ranknow', 'name', 'market'])]
crp2 = crp2[crp2['date']>='2018-01-01']
crp2[['ranknow','name', 'market']].groupby('name').describe()
crp2 = crp2.sort_values('ranknow')
crp2[['name', 'market']].boxplot(by='name', figsize=(10,20))
```

```python
# Scatter plot Bitcoin vs Etheruem

crp5 = crp[crp['ranknow']<=5]
crp5 = crp5[(['ranknow', 'date', 'market'])]

crp_1 = crp5[(crp5['ranknow']==1) & (crp5['date']>='2018-01-01') ]
crp_1 = crp_1['market']
crp_2 = crp5[(crp5['ranknow']==2) & (crp5['date']>='2018-01-01') ]
crp_2 = crp_2['market']
crp_5 = crp5[(crp5['ranknow']==5) & (crp5['date']>='2018-01-01') ]
crp_5 = crp_5['market']

plt.figure(figsize=(15,5))

plt.subplot(1,2,1)
plt.scatter(x=crp_1, y=crp_2)
plt.title('Bitcoin vs. Ethereum')
plt.xlabel('BTC')
plt.ylabel('ETH')

plt.subplot(1,2,2)
plt.scatter(x=crp_1, y=crp_5)
plt.title('Bitcoin vs. EOS')
plt.xlabel('BTC')
plt.ylabel('EOS')
```
