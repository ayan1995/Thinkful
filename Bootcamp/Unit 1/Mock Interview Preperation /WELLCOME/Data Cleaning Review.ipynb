```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv('WELLCOME_APCspend2013_forThinkful.csv', encoding = 'latin_1')
df
```

```python
# Data Cleaning

df['COST (£) charged to Wellcome (inc VAT when charged)'] = df['COST (£) charged to Wellcome (inc VAT when charged)'].str.replace('£', '')
df['COST (£) charged to Wellcome (inc VAT when charged)'] = df['COST (£) charged to Wellcome (inc VAT when charged)'].str.replace('$', '')
df = df.rename(index=str, columns={"COST (£) charged to Wellcome (inc VAT when charged)": "COST" })
df['COST'] = pd.to_numeric(df['COST'])
df = df[df['COST'] < 10000.00]
```

```python
g = sns.distplot(df.COST)
```

```python
df['Publisher'] = df['Publisher'].str.capitalize()
df['Publisher'] = df['Publisher'].str.strip()
df['Publisher'] = df['Publisher'].replace({'Cold Spring Harbor Publications':'Cold Spring Harbor Press', 'Cold Spring Harbor Laboratory Press': 'Cold Spring Harbor Press', 'Cold Spring Harbor': 'Cold Spring Harbor Press', 'Cenveo Publisher Services/ASM JV1': 'Cenveo Publisher Services', 'Wiley-vch': 'Wiley-blackwell','Cambridge Univ Press': 'Cambridge University Press', 'Cambridge Uni Press': 'Cambridge University Press', 'BMJ Publishing Group Ltd': 'BMJ Publishing Group', 'BMJ Journals': 'BMJ Group', 'Wolters kluwer health': 'Wolters kluwer', 'Wolters kluwers': 'Wolters kluwer', 'Wolters kluwer n.v./lippinott': 'Wolters kluwer', 'Acs publications': 'Acs (amercian chemical society) publications', 'Wiley-blackwell, john wiley & sons': 'Wiley-blackwell', 'Wliey-blackwell': 'Wiley-blackwell'})
```
