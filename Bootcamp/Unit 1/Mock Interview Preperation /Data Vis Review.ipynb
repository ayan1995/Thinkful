```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
sns.set(style = 'dark')
```

```python
tips = sns.load_dataset("tips")
```

```python
tips
```

```python
g = sns.boxplot(x = 'sex', y = 'tip', data = tips)
plt.show()

g = sns.factorplot(x = 'sex', y = 'tip', data = tips, hue = 'time', kind = 'bar', ci = 95)
plt.show()

g = sns.factorplot(x = 'sex', y = 'tip', data = tips, hue = 'time', kind = 'point', ci = 95, dodge = True, join = False)
plt.show()

g = sns.catplot(x = 'day', y = 'tip', data = tips, hue = 'sex')
plt.show()
```

```python
diamonds = sns.load_dataset('diamonds')
diamonds
```

```python
g = sns.catplot(x = 'color', y = 'price', data = diamonds, kind = 'boxen')
```

```python
g = sns.catplot(x = 'total_bill', y = 'day', data = tips, kind = 'violin')
```

```python
g = sns.lmplot(x = 'tip', y = 'total_bill', data = tips, hue = 'time')

g = sns.relplot(x = 'tip', y = 'total_bill', data = tips, col = 'time', hue = 'sex', kind = 'scatter')

df = pd.DataFrame(dict(time=pd.date_range("2017-1-1", periods=500),
                       value=np.random.randn(500).cumsum()))

g = sns.relplot(x = 'time', y = 'value', data = df, kind = 'line')
g.fig.autofmt_xdate()
```

```python
fmri = sns.load_dataset('fmri')
sns.relplot(x="timepoint", y="signal", hue="subject",
            col="region", row="event", height=3,
            kind="line", estimator=None, data=fmri);
```

```python
sns.relplot(x="timepoint", y="signal", hue="event", style="event",
            col="subject", col_wrap=5,
            height=3, aspect=.75, linewidth=2.5,
            kind="line", data=fmri.query("region == 'parietal'"))
```
