```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import scipy.stats as stats
import seaborn as sns
%matplotlib inline
```

```python
df_national = pd.read_csv('MPI_national.csv')
df_subnational = pd.read_csv('MPI_subnational.csv')
```

## Choose one variable and plot that variable four different ways.

```python
df_subnational
```

```python
df_ubran_deprivation = df_national.head(20)
df_urban_deprivation = df_ubran_deprivation[(['Country', 'Intensity of Deprivation Urban'])]

ax = sns.barplot(data = df_urban_deprivation.reset_index(), x = 'index', y = 'Intensity of Deprivation Urban')
plt.show()

ax = sns.pointplot(data = df_urban_deprivation.reset_index(), x = 'index', y = 'Intensity of Deprivation Urban', join = False)
plt.show()

ax = sns.distplot(df_ubran_deprivation['Intensity of Deprivation Urban'])
plt.show()

ax = sns.lineplot(data = df_urban_deprivation.reset_index(), x = 'index', y = 'Intensity of Deprivation Urban')
plt.show()
```

## Choose two continuous variables, and plot them three different ways.

```python
headcount_vs_Intensity = df_national[(['Headcount Ratio Urban', 'Intensity of Deprivation Urban'])]
x = headcount_vs_Intensity['Headcount Ratio Urban']
y = headcount_vs_Intensity['Intensity of Deprivation Urban']

ax = sns.regplot(x, y)
plt.show()

ax = sns.lineplot(x, y)
plt.show()

ax = sns.lmplot('Headcount Ratio Urban', 'Intensity of Deprivation Urban', data = headcount_vs_Intensity)
plt.show()

```

## Choose one continuous variable and one categorical variable, and plot them six different ways.

```python
df_ubran_deprivation = df_national.head(20)
df_urban_deprivation = df_ubran_deprivation[(['Country', 'Intensity of Deprivation Urban'])]
x = df_urban_deprivation['Country']
y = df_urban_deprivation['Intensity of Deprivation Urban']

ax = sns.barplot(x, y)
ax.set_xticklabels(ax.get_xticklabels(),rotation=90)
plt.show()

ax = sns.pointplot(x, y, join=False)
ax.set_xticklabels(ax.get_xticklabels(),rotation=90)
plt.show()

ax = sns.lineplot(x, y)
ax.set_xticklabels(ax.get_xticklabels(),rotation=90)
plt.show()

ax = sns.scatterplot(x, y)
ax.set_xticklabels(ax.get_xticklabels(),rotation=90)
plt.show()

Afghanisan_urban_deprivation = df_subnational[df_subnational['Country'] == 'Afghanistan']
Afghanisan_urban_deprivation = Afghanisan_urban_deprivation['Intensity of deprivation Regional']
y = Afghanisan_urban_deprivation 

ax = sns.boxplot(y)
ax.set_title('Range of Intensity of deprivation Regional for Afghanistan')
plt.show()

ax = sns.distplot(y)
ax.set_title('Distribution of Intensity of deprivation Regional for Afghanistan')
plt.show()
```
